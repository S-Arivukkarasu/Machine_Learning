```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import sklearn
```

```python
# Importing data from csv files
raw_data = pd.read_csv("https://docs.google.com/spreadsheets/d/1JTivlfWdkA_auM4kof-FKWCsjB6OpHVhztqSORvhirs/export?gid=2002571320&format=csv")
raw_data
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1, 'No': 0})
```

```python
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(data['SAT'], y )
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
fig = plt.plot(x1,yhat_no, lw=2, c='red')
fig = plt.plot(x1,yhat_yes, lw=2, c='green')
plt.xlabel("SAT Score", fontsize=20)
plt.ylabel("GPA", fontsize=20)
plt.xlim(1500)
plt.show()
```

```python
plt.scatter(data['SAT'], y , c=data['Attendance'], cmap='RdYlBu_r')
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
fig = plt.plot(x1,yhat_no, lw=2, c='red')
fig = plt.plot(x1,yhat_yes, lw=2, c='green')
plt.xlabel("SAT Score", fontsize=20)
plt.ylabel("GPA", fontsize=20)
plt.xlim(1600)
plt.show()
```
