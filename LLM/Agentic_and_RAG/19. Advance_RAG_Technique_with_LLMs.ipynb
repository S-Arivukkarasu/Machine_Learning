# Advanced RAG Techniques with LLMs

Retrieval-Augmented Generation (RAG) enhances Large Language Models by combining them with external knowledge sources. Instead of relying only on the model’s training data, RAG retrieves relevant documents from a knowledge base and uses them as context when generating answers.

In this notebook, we build an **Advanced RAG pipeline** that improves retrieval quality, relevance, and answer accuracy using multiple techniques such as intelligent chunking, reranking, and query rewriting.

## Pipeline Overview

The complete workflow of this Advanced RAG system:

1. Load documents from a knowledge base
2. Use an LLM to create structured document chunks
3. Generate embeddings for each chunk
4. Store embeddings in a vector database (ChromaDB)
5. Visualize embeddings using dimensionality reduction
6. Rewrite user queries for better retrieval
7. Retrieve top-K relevant chunks
8. Rerank chunks using an LLM
9. Generate a final response using retrieved context

This approach improves both **retrieval accuracy** and **answer quality** compared to basic RAG systems.

## Techniques Used in This Advanced RAG System

### 1. Intelligent Document Chunking
Instead of simple text splitting, an LLM is used to divide documents into meaningful semantic chunks. Each chunk contains:
- **Headline** – a short title for the chunk
- **Summary** – a brief explanation of the content
- **Original Text** – the exact text from the document

This helps improve semantic retrieval and context preservation.

### 2. Overlapping Chunks
Chunks are generated with about **25% overlap** between neighboring sections.  
This ensures important information is not lost at chunk boundaries and improves retrieval quality.

### 3. Vector Embeddings and Storage
Each chunk is converted into vector embeddings using an embedding model and stored in **ChromaDB**, a vector database that allows efficient similarity search.

### 4. Vector Space Visualization
Embeddings are visualized using **t-SNE dimensionality reduction** to create:
- 2D visualization
- 3D visualization

This helps understand how documents cluster semantically in embedding space.

### 5. Query Rewriting
User questions are rewritten into a **short and focused search query** before retrieval. This improves the chances of retrieving the most relevant knowledge base content.

### 6. LLM-Based Reranking
After retrieving the top-K chunks, an LLM evaluates their relevance and **reranks them from most relevant to least relevant**.

### 7. Context-Aware Answer Generation
The final response is generated using:
- The reranked chunks
- The conversation history
- A system prompt guiding the assistant

This ensures answers are **accurate, relevant, and grounded in the retrieved knowledge base context**.

```python
# Importing Libraries

from pathlib import Path
from ollama import embed
from pydantic import BaseModel, Field
from chromadb import PersistentClient
from tqdm import tqdm
from litellm import completion
import numpy as np
from sklearn.manifold import TSNE
import plotly.graph_objects as go
```

```python
# Model Name and other details

MODEL_ANSWER = "ollama/llama3.1:latest" # dont forget to pull the Models before using it 
MODEL_JUDGE = "ollama/mistral:latest"
MODEL_CHUNKS = "ollama/gemma3:latest"

DB_NAME = "/home/alexender/Desktop/Projects/My_projects/Data/prod-ready-rag"
collection_name = "advance_rag_docs"
embedding_model = "nomic-embed-text-v2-moe"
KNOWLEDGE_BASE_PATH = Path("/home/alexender/Desktop/Projects/My_projects/Data/knowledge-base")
AVERAGE_CHUNK_SIZE = 500
```

```python
# Inspired by LangChain's Document - let's have something similar

class Result(BaseModel):
    page_content: str
    metadata: dict
```

```python
# A class to perfectly represent a chunk

class Chunk(BaseModel):
    headline: str = Field(description="A brief heading for this chunk, typically a few words, that is most likely to be surfaced in a query")
    summary: str = Field(description="A few sentences summarizing the content of this chunk to answer common questions")
    original_text: str = Field(description="The original text of this chunk from the provided document, exactly as is, not changed in any way")

    def as_result(self, document):
        metadata = {"source": document["source"], "type": document["type"]}
        return Result(page_content=self.headline + "\n\n" + self.summary + "\n\n" + self.original_text,metadata=metadata)


class Chunks(BaseModel):
    chunks: list[Chunk]
```

### Three steps for Ingest Part:
1. Fetch documents from the knowledge base, like LangChain did
2. Call an LLM to turn documents into Chunks
3. Store the Chunks in Chroma

#### Let's start with Step 1

```python
def fetch_documents():
    """A homemade version of the LangChain DirectoryLoader"""

    documents = []

    for folder in KNOWLEDGE_BASE_PATH.iterdir():
        doc_type = folder.name
        for file in folder.rglob("*.md"):
            with open(file, "r", encoding="utf-8") as f:
                documents.append({"type": doc_type, "source": file.as_posix(), "text": f.read()})

    print(f"Loaded {len(documents)} documents")
    return documents
```

#### Donezo! On to Step 2 - make the chunks

```python
def make_prompt(document):
    how_many = (len(document["text"]) // AVERAGE_CHUNK_SIZE) + 1
    return f"""
You take a document and you split the document into overlapping chunks for a KnowledgeBase.

The document is from the shared drive of a company called Insurellm.
The document is of type: {document["type"]}
The document has been retrieved from: {document["source"]}

A chatbot will use these chunks to answer questions about the company.
You should divide up the document as you see fit, being sure that the entire document is returned in the chunks - don't leave anything out.
This document should probably be split into {how_many} chunks, but you can have more or less as appropriate.
There should be overlap between the chunks as appropriate; typically about 25% overlap or about 50 words, so you have the same text in multiple chunks for best retrieval results.

For each chunk, you should provide a headline, a summary, and the original text of the chunk.
Together your chunks should represent the entire document with overlap.

Here is the document:

{document["text"]}

Respond with the chunks.
"""
```

```python
# print(make_prompt(documents[0]))
```

```python
def make_messages(document):
    return [
        {"role": "user", "content": make_prompt(document)},
    ]
```

```python
# make_messages(documents[0])
```

```python
def process_document(document):
    messages = make_messages(document)
    response = completion(model=MODEL_CHUNKS, messages=messages, response_format=Chunks)
    reply = response.choices[0].message.content
    doc_as_chunks = Chunks.model_validate_json(reply).chunks
    return [chunk.as_result(document) for chunk in doc_as_chunks]
```

```python
# print(process_document(documents[0]))
```

```python
def create_chunks(documents):
    chunks = []
    for doc in tqdm(documents):
        chunks.extend(process_document(doc))
    return chunks
```

#### Finally, Step 3 - save the embeddings

```python
def create_embeddings(chunks):
    chroma = PersistentClient(path=DB_NAME)
    if collection_name in [c.name for c in chroma.list_collections()]:
        chroma.delete_collection(collection_name)

    texts = [chunk.page_content for chunk in chunks]
    emb = embed(model=embedding_model, input=texts)
    vectors = emb["embeddings"]

    collection = chroma.get_or_create_collection(collection_name)

    ids = [str(i) for i in range(len(chunks))]
    metas = [chunk.metadata for chunk in chunks]

    collection.add(ids=ids, embeddings=vectors, documents=texts, metadatas=metas)
    print(f"Vectorstore created with {collection.count()} documents")
```

```python
create_embeddings(chunks)
```

### And now - let's build an Advanced RAG!

We will use these techniques:

1. Reranking - reorder the rank results
2. Query re-writing


```python
class RankOrder(BaseModel):
    order: list[int] = Field(
        description="The order of relevance of chunks, from most relevant to least relevant, by chunk id number"
    )
```

```python
def rerank(question, chunks):
    system_prompt = """
You are a document re-ranker.
You are provided with a question and a list of relevant chunks of text from a query of a knowledge base.
The chunks are provided in the order they were retrieved; this should be approximately ordered by relevance, but you may be able to improve on that.
You must rank order the provided chunks by relevance to the question, with the most relevant chunk first.
Reply only with the list of ranked chunk ids, nothing else. Include all the chunk ids you are provided with, reranked.
"""
    user_prompt = f"The user has asked the following question:\n\n{question}\n\nOrder all the chunks of text by relevance to the question, from most relevant to least relevant. Include all the chunk ids you are provided with, reranked.\n\n"
    user_prompt += "Here are the chunks:\n\n"
    for index, chunk in enumerate(chunks):
        user_prompt += f"# CHUNK ID: {index + 1}:\n\n{chunk.page_content}\n\n"
    user_prompt += "Reply only with the list of ranked chunk ids, nothing else."
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt},
    ]
    response = completion(model=MODEL_JUDGE, messages=messages, response_format=RankOrder)
    reply = response.choices[0].message.content
    order = RankOrder.model_validate_json(reply).order
    print(order)
    return [chunks[i - 1] for i in order]
```

```python
RETRIEVAL_K = 10

def fetch_context_unranked(question):
    query = embed(model=embedding_model, input=[question])["embeddings"][0]
    results = collection.query(query_embeddings=[query], n_results=RETRIEVAL_K)
    chunks = []
    for result in zip(results["documents"][0], results["metadatas"][0]):
        chunks.append(Result(page_content=result[0], metadata=result[1]))
    return chunks
```

```python
def fetch_context(question):
    chunks = fetch_context_unranked(question)
    return rerank(question, chunks)
```

```python
SYSTEM_PROMPT = """
You are a knowledgeable, friendly assistant representing the company Insurellm.
You are chatting with a user about Insurellm.
Your answer will be evaluated for accuracy, relevance and completeness, so make sure it only answers the question and fully answers it.
If you don't know the answer, say so.
For context, here are specific extracts from the Knowledge Base that might be directly relevant to the user's question:
{context}

With this context, please answer the user's question. Be accurate, relevant and complete.
"""
```

```python
# In the context, include the source of the chunk

def make_rag_messages(question, history, chunks):
    context = "\n\n".join(f"Extract from {chunk.metadata['source']}:\n{chunk.page_content}" for chunk in chunks)
    system_prompt = SYSTEM_PROMPT.format(context=context)
    return [{"role": "system", "content": system_prompt}] + history + [{"role": "user", "content": question}]
```

```python
def rewrite_query(question, history=[]):
    """Rewrite the user's question to be a more specific question that is more likely to surface relevant content in the Knowledge Base."""
    message = f"""
You are in a conversation with a user, answering questions about the company Insurellm.
You are about to look up information in a Knowledge Base to answer the user's question.

This is the history of your conversation so far with the user:
{history}

And this is the user's current question:
{question}

Respond only with a single, refined question that you will use to search the Knowledge Base.
It should be a VERY short specific question most likely to surface content. Focus on the question details.
Don't mention the company name unless it's a general question about the company.
IMPORTANT: Respond ONLY with the knowledgebase query, nothing else.
"""
    response = completion(model=MODEL_CHUNKS, messages=[{"role": "system", "content": message}])
    return response.choices[0].message.content
```

```python
rewrite_query("Who won the IIOTY award?", [])
```

```python
def answer_question(question: str, history: list[dict] = []) -> tuple[str, list]:
    """
    Answer a question using RAG and return the answer and the retrieved context
    """
    query = rewrite_query(question, history)
    print(query)
    chunks = fetch_context(query)
    messages = make_rag_messages(question, history, chunks)
    response = completion(model=MODEL_ANSWER, messages=messages)
    return response.choices[0].message.content, chunks
```
