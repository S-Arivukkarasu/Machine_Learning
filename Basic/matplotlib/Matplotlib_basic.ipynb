## Matplotlib workflow examples

```python
# 0. Import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# Prepare some Data
x = [1,2,3,4,5]
y = [11, 22, 33, 44, 55]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # width, height

# 3. Plot figure
ax.plot(x, y)

# 4. Customize the plot
ax.set(title="Simple Plot",
              xlabel="x-axis",
              ylabel="y-axis")

# 5. save figure
fig.savefig("images/simple_plot.png")
```

## Making Plots from NumPy array

* Line Plot

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Line Plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Scatter Plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig ,ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make plots from dictionary Bar Plots
nut_prices = {"Almod_nuts": 10,
             "Peanut_nuts": 8,
             "cashew_nuts": 23}
fig, ax = plt.subplots()
ax.bar(x=list(nut_prices.keys()), height=nut_prices.values());
ax.set(title="Nut Shop", ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_prices.keys()), list(nut_prices.values()));
```

```python
# Histograms
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Subplot Options

```python
# Subplot Option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                                                               ncols=2,
                                                                                figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(100), np.random.random(100));
ax3.bar(nut_prices.keys(), nut_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot Option 2

fig, ax = plt.subplots(nrows=2,
                                        ncols=2,
                                        figsize=(10, 5))
ax[0, 0].plot(x, x/2);
ax[0,1].scatter(np.random.random(100), np.random.random(100));
ax[1,0].bar(nut_prices.keys(), nut_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
# New Data Set - Heart Diseases dataset

heart_diesase = pd.read_csv("https://docs.google.com/spreadsheets/d/1Cm_sP6krAIxTbg-G3JWavWUcQyed3IITbFbttlXIh9k/export?gid=1886998129&format=csv")
heart_diesase.head()
```

```python
# creating hist plot for age 
heart_diesase["age"].plot.hist();
```

```python
heart_diesase.plot.hist(figsize=(10, 20), subplots=True);
```

```python
heart_diesase.head()
```

```python
over_50 = heart_diesase[heart_diesase["age"]>50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter",
                            x="age", y="chol", c="target")
```

```python
## oo method

fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot.scatter(x="age", y="chol", c='target', ax=ax);
# set x limiters
ax.set_xlim([0, 100]);
```

```python
# Pure oo method
fig, ax = plt.subplots(figsize=(10,6))
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["red","green"])
scatter = ax.scatter(x=over_50["age"],
                                       y=over_50["chol"],
                                        c=over_50["target"],
                                        cmap=cmap);
#Customizing the plot
ax.set(title="Heart Disease and Cholestral Levels",
              xlabel="Age", ylabel="Cholestrol");
#Adding Legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add a line horizontally
ax.axhline(over_50["chol"].mean(),
                      linestyle="--");

plt.show();
```

```python
# Add subplots

fig , (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)

# Add data to ax0
cmap0 = matplotlib.colors.LinearSegmentedColormap.from_list("", ["red","blue"])
scatter=ax0.scatter(x=over_50["age"], 
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap=cmap0)
ax0.set(title="Heart Disease vs Cholestrol", xlabel="Age", ylabel="Cholestral");
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(over_50["chol"].mean(), linestyle="--")

#Add data to ax1
cmap1 = matplotlib.colors.LinearSegmentedColormap.from_list("", ["green","yellow"])
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap=cmap1)
ax1.set(title="Heart Disease vs Max Heart Beat Rate", xlabel="Age", ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle="--")

# Add title to the whole figure
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold")

plt.show();


```
