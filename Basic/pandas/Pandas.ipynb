## Manipulating Data

```python
car_sales_missing = pd.read_csv('https://docs.google.com/spreadsheets/d/1gm_T24Uj0v6ZBipYBy1_-2mp8ffsm6mgmrS2dN5smuc/export?gid=1239395861&format=csv')
car_sales_missing
```

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), 
                                    inplace=True)
car_sales_missing
```

```python
car_sales_missing.dropna()
```

```python
car_sales
```

```python
# column from Series
seats = pd.Series([5,5,5,5,5,4,4,4,4,4])

car_sales["Seats"] = seats
car_sales
```

```python
# cloumn from List
fuel = [7.5, 6.4, 9.2, 5.0, 9.6, 3.7, 4.3, 5.9, 6.5, 8.2]

car_sales["Fuel per 100KM"] = fuel
car_sales
```

```python
car_sales["Total fuel (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
# column from a single value

car_sales["Number of Wheels"] = 4
car_sales
```
